```python
import pandas as pd

# Assuming you have a DataFrame called 'table'
tsv_file_path = 'amr_counts_collapsed_20230605.tsv'

# Load the TSV file into a DataFrame
table = pd.read_csv(tsv_file_path, sep='\t')
print(table)
```

```python


# Step 1: Identify the label column
label_column = table.iloc[:, 0]  # Assuming the label column is the first column
print(label_column)
```

```python

# Step 2: Copy the label column to a separate list or array
labels = label_column.tolist()
print(labels)
```

```python

# Step 3: Sort each column in descending order, while keeping track of the corresponding labels
sorted_table = table.apply(lambda col: col.sort_values(ascending=False)[::-1], axis=0)
print(sorted_table)
```

```python

# Step 4: Replace the values in each column with the corresponding labels
sorted_table = sorted_table.apply(lambda col: labels, axis=0)
type(sorted_table)
# Display the sorted table with labels
#print(sorted_table)
```

```python
csv_file_path = 'sorted_file_amr.csv'

# Convert the DataFrame to a CSV file
sorted_table.to_csv(csv_file_path, index=False)
```
